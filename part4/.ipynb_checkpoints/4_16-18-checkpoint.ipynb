```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('시도별 전출입 인구수.xlsx')
```

```python
df.head()
```

```python
df = df.fillna(method='ffill')
```

```python
mask = (df['전출지별'] == '서울특별시') 
df_seoul = df[mask]
```

```python
df_seoul = df_seoul.drop(['전출지별'], axis=1)
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul.head()
```

### [예제]4_16

```python
col_years = list(map(str,range(2010,2018)))
```

```python
df4 = df_seoul.loc[['충청남도', '경상북도','강원도','전라남도'], col_years]
df4 = df4.T
```

```python
df4
```

```python
plt.style.use('ggplot')

df4.index = df4.index.map(int)

df4.plot(kind='bar', figsize=(20,10), width=0.7, color=['orange','green','skyblue','blue'])

plt.title('서울 -> 타시도 인구 이동', size=30)
plt.ylabel('이동 인구 수', size=20)
plt.xlabel('기간', size=20)
plt.ylim(5000,30000)
plt.legend(fontsize=15)

plt.show()
```

### [예제]4_17

```python
df4 = df_seoul.loc[['충청남도', '경상북도','강원도','전라남도'], col_years]
df4['합계'] = df4.sum(axis=1)
```

```python
df4
```

```python
df_total = df4[['합계']].sort_values(by='합계')
df_total
```

```python
plt.style.use('ggplot')

df_total.plot(kind='barh', color='cornflowerblue', width=0.5, figsize=(10,5))

plt.title('서울 -> 타시도 인구 이동')
plt.ylabel('전입지', rotation=90)
plt.xlabel('이동 인구 수')

plt.show()
```

### [예제]4_18

```python
from matplotlib import font_manager, rc
```

```python
font_path = "data/malgun.ttf"
```

```python
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
plt.style.use('ggplot')
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_excel('data/남북한발전전력량.xlsx', convert_float=True)
df = df.loc[5:9]
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop('전력량 (억㎾h)', axis='columns', inplace=True)
df.set_index('발전 전력별', inplace=True)
df = df.T
```

```python
df = df.rename(columns={'합계':'총발전량'})
df['총발전량 - 1년'] = df['총발전량'].shift(1)
df['증감률'] = ((df['총발전량']/df['총발전량 - 1년']) - 1)*100
```

```python
ax1 = df[['수력','화력']].plot(kind='bar', figsize=(20,10), width=0.7, stacked=True)
ax2 = ax1.twinx()
ax2.plot(df.index, df.증감률, ls='--', marker='o', ms=20, color='green', label='전년대비 증감률(%)')

ax1.set_ylim(0,500)
ax2.set_ylim(-50,50)

ax1.set_xlabel('연도', size=20)
ax1.set_ylabel('발전량 (억 KWh)')
ax2.set_ylabel('전년 대비 증감률(%)')

plt.title('북한 전력 발전량 (1990~2016)', size=30)
ax1.legend(loc='upper left')

plt.show()
```
