```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import platform

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':  # 맥OS 
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':  # 윈도우
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...  sorry~~~')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('시도별 전출입 인구수.xlsx')
df
```

```python
df.head()
```

```python
df = df.fillna(method='ffill')
df.head()
```

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
df_seoul = df[mask]
df_seoul
```

```python
df_seoul = df_seoul.drop(['전출지별'], axis=1)
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
```

```python
df_seoul.set_index('전입지', inplace=True)
```

```python
col_years = list(map(str, range(1970,2018)))
```

```python
df4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],col_years]
df4 = df4.T
```

```python
df4.head()
```

```python
plt.style.use('ggplot')
```

```python
df4.index = df4.index.map(int)
```

```python
df4.head()
```

```python
df4.plot(kind='area', stacked=False, alpha=0.2, figsize=(20,10))

plt.title('서울 -> 타시도 인구 이동', size=30)
plt.ylabel('이동 인구 수 ', size=20)
plt.xlabel('기간', size=20)
plt.legend(fontsize=10)

plt.show()
```

```python
df4.plot(kind='area', stacked=True, alpha=0.2, figsize=(20,10))

plt.title('서울 -> 타시도 인구 이동', size=30)
plt.ylabel('이동 인구 수 ', size=20)
plt.xlabel('기간', size=20)
plt.legend(fontsize=10)

plt.show()
```

```python
ax = df4.plot(kind='area', stacked=True, alpha=0.2, figsize=(20,10))
print(type(ax))

ax.set_title('서울 -> 타시도 인구 이동', size=30, color='brown', weight='bold')
ax.set_ylabel('이동 인구 수 ', size=20, color='blue')
ax.set_xlabel('기간', size=20, color='blue')
ax.legend(fontsize=10)

plt.show()
```
